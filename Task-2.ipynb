```python
# Hangman Game
import random

# Words and hints
word_list = [
    {"word": "python", "hint": "A popular programming language"},
    {"word": "elephant", "hint": "The largest land animal"},
    {"word": "guitar", "hint": "A musical instrument with strings"},
    {"word": "volcano", "hint": "A mountain that erupts"},
    {"word": "hangman", "hint": "This game you're playing now!"}
]

# ASCII Art for Hangman Stages
hangman_stages = [
    """
     ------
     |    |
     |    O
     |   /|\\
     |   / \\
     |
    ---------
    """,
    """
     ------
     |    |
     |    O
     |   /|\\
     |   /
     |
    ---------
    """,
    """
     ------
     |    |
     |    O
     |   /|\\
     |
     |
    ---------
    """,
    """
     ------
     |    |
     |    O
     |   /|
     |
     |
    ---------
    """,
    """
     ------
     |    |
     |    O
     |    |
     |
     |
    ---------
    """,
    """
     ------
     |    |
     |    O
     |
     |
     |
    ---------
    """,
    """
     ------
     |    |
     |
     |
     |
     |
    ---------
    """
]

def play_hangman():
    entry = random.choice(word_list)
    word = entry["word"]
    hint = entry["hint"]
    guessed = ['_'] * len(word)
    guessed_letters = []
    tries = len(hangman_stages) - 1

    print("üéÆ Welcome to Hangman!")
    print(f"üí° Hint: {hint}")

    while tries > 0 and '_' in guessed:
        print(hangman_stages[tries])
        print("Word: " + ' '.join(guessed))
        print(f"Guessed letters: {', '.join(guessed_letters)}")

        guess = input("Enter a letter: ").lower()

        if len(guess) != 1 or not guess.isalpha():
            print("‚ö†Ô∏è Enter a single alphabet letter.")
            continue
        if guess in guessed_letters:
            print("üîÅ You already guessed that letter.")
            continue

        guessed_letters.append(guess)

        if guess in word:
            for i, char in enumerate(word):
                if char == guess:
                    guessed[i] = guess
            print("‚úÖ Correct!")
        else:
            tries -= 1
            print("‚ùå Wrong guess.")

    # Game over message
    if '_' not in guessed:
        print("üéâ You won! The word was:", word)
    else:
        print(hangman_stages[tries])
        print("üíÄ Game over. The word was:", word)

# Run the game
if __name__ == "__main__":
    play_hangman()


```

```python
'''Web Scraper: Extract data from websites using libraries like
Beautiful Soup or Scrapy.'''
# .   


import requests
from bs4 import BeautifulSoup

# Step 1: Send an HTTP request
url = 'https://timesofindia.indiatimes.com/'  # Replace with your target URL
response = requests.get(url)

# Step 2: Parse the page content with Beautiful Soup
soup = BeautifulSoup(response.text, 'html.parser')

# Step 3: Find and extract data
# For example, extracting headlines inside <h2> tags
headlines = soup.find_all('h2')

# Step 4: Print the extracted headlines
for i, headline in enumerate(headlines, 1):
    print(f"{i}. {headline.text.strip()}")
```
